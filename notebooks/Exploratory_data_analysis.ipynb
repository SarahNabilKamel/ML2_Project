```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_20newsgroups
```

```python
# ðŸ”¹ load 20 Newsgroups
def load_newsgroups(categories=None):
    print("ðŸ“¥ Loading 20 Newsgroups dataset...")
    newsgroups = fetch_20newsgroups(subset='all', categories=categories, remove=('headers', 'footers', 'quotes'))
    df = pd.DataFrame({'text': newsgroups.data, 'category': newsgroups.target})
    df['category_name'] = df['category'].apply(lambda x: newsgroups.target_names[x])
    return df

# ðŸ”¹ load Wikipedia People
def load_people_wiki():
    print("ðŸ“¥ Loading People Wikipedia dataset...")
    df = pd.read_csv('C:/Users/Sarah Kamel/Desktop/9-month/ML2/Ml2-Project/data/Dataset/people_wiki.csv')
    return df[['name', 'text']]
```

```python
# ðŸ“Š data analysis 20 Newsgroups
df_news = load_newsgroups()
print(df_news.info())
print(df_news.head())
print("\nðŸ”Ž Missing values:")
print(df_news.isnull().sum())

plt.figure(figsize=(12, 5))
sns.countplot(y=df_news['category_name'], order=df_news['category_name'].value_counts().index)
plt.title("Distribution of Categories in 20 Newsgroups")
plt.show()

# ðŸ“Š  data analysis Wikipedia People
df_wiki = load_people_wiki()
print(df_wiki.info())
print(df_wiki.head())
print("\nðŸ”Ž Missing values:")
print(df_wiki.isnull().sum())

plt.figure(figsize=(8, 5))
plt.hist(df_wiki['text'].str.len(), bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Text Lengths in Wikipedia People Dataset")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()
```
